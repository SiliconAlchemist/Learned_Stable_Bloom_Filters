```python
file = open("dataset/malicious_urls.txt", "r",encoding='utf-8')

ctr1 = 0
ctr2 = 0

X = []
y = []

for i, line in enumerate(file.readlines()[1:]):
    url = line[:-5]
    label = line[-5:-1]
    
    url.strip(',')
    url.strip("")
    
    if ((ctr1==30000) and (ctr2==30000)):
        break
    if label==",bad":
        if ctr1<30000:
            ctr1+=1
            X.append(url)
            y.append(0)
    else:
        if ctr2<30000:
            ctr2+=1
            X.append(url)
            y.append(1)
        
data = []
for i in range(len(X)):
    data.append([X[i], y[i]])
data
```
