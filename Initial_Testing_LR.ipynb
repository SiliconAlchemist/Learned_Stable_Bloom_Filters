```python
import numpy as np
import pandas as pd
import sklearn
import joblib
from stable_bloom_filter import stable_bloom_filter
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
data = pd.read_csv("Dataset/train.csv")
```

```python
data.head()
```

```python
data_processed = {'Action':[], 'Employee_Det':[]}
```

```python
for ind in range(len(data)):
    det = ''.join(map(str, list(data.iloc[ind,1:])))
    act = data.iloc[ind,0]

    data_processed['Action'].append(act)
    data_processed['Employee_Det'].append(det)
data_new = pd.DataFrame(data_processed)
data_new.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_new['Employee_Det'], 
                                                    data_new['Action'], test_size=0.30, 
                                                    )
```

```python
y_train[0:5]
```

```python
X_train=np.asarray(X_train)
y_train=np.asarray(y_train)
#X_train=X_train.reshape(X_train.shape[0],1)
#y_train=y_train.reshape(y_train.shape[0],1)
X_test=np.asarray(X_test)
y_test=np.asarray(y_test)
#X_test=X_test.reshape(X_test.shape[0],1)
#y_test=y_test.reshape(y_test.shape[0],1)
```

```python
y_train.shape
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
joblib.dump(logmodel, 'm1.pkl')
```

```python
print(classification_report(y_test,predictions)
```
